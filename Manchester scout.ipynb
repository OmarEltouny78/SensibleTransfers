```python
compare_df=final_df[["Rk","age_bins","TklTackles","TklWTackles","Def 3rdTackles","Mid 3rdTackles","Tkl%VsD","%Press","Def 3rdPress","Blocks",'Cmp%T', 'TotDistT', 'PrgDistT','Cmp%S','Cmp%M','Cmp%L','1/3', 'Prog']]
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
y=le.fit_transform(compare_df.age_bins)
```

```python
compare_df['Age']=y
```

```python
compare_df.drop(columns=["age_bins"],inplace=True)
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
wcss=[]
for i in range(1,7):
    kmeans = KMeans(i)
    kmeans.fit(compare_df)
    wcss_iter = kmeans.inertia_
    wcss.append(wcss_iter)

number_clusters = range(1,7)
plt.plot(number_clusters,wcss)
plt.title('The Elbow title')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
```

```python
kmeans = KMeans(3)
kmeans.fit(compare_df)
```

```python
identified_clusters = kmeans.fit_predict(compare_df)
identified_clusters
```

```python
compare_df["cluster"]=identified_clusters
compare_df.head()
compare_df.cluster.value_counts()
```

```python
filtered_label=compare_df[compare_df.cluster==0]
filtered_label2=compare_df[compare_df.cluster==1]
filtered_label3=compare_df[compare_df.cluster==2]
filtered_label.head()
plt.scatter(filtered_label.iloc[:,0] , filtered_label.iloc[:,1])
plt.scatter(filtered_label2.iloc[:,0] , filtered_label2.iloc[:,1] , color = 'red')
plt.scatter(filtered_label3.iloc[:,0] , filtered_label3.iloc[:,1] , color = 'green')
plt.show()
```
